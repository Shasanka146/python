```python
import matplotlib.pyplot as plt
labels=['Group1','Group2','Group3','Group4','Group5']
men_age=[20,25,30,35,27]
women_age=[25,32,34,20,25]
fig, ax=plt.subplots()
ax.bar(labels,men_age,label='Men')
ax.bar(labels,women_age,bottom=men_age,label="Women")
ax.set_xlabel('Group Number')
ax.set_title("Gender Ratio by Group")
ax.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
classes=["Python","R","AI","ML","DS"]
class1_Students=[30,10,20,25,10]
class2_Students=[40,5,20,20,10]
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_Students,width,label="class1 Students")
plt.bar(class_index,class2_Students,width,label="class2 Students")
plt.xticks(class_index+width,classes)

plt.title("Bar charts of classes")
plt.xlabel("classes")
plt.ylabel("no. of students")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as num
#line plot
x_axis=num.array([0,5,10,15,20,25,30])
plt.plot(x_axis,x_axis*2)
plt.ylabel("Yaxis")
plt.xlabel("Xlabel")
plt.title("My first plot")
```

```python
plt.plot(x_axis,x_axis*2*2,'red',label='x=y**2')
plt.legend()
plt.show()
```

```python

empty_canvas = plt.figure()
y_axis = x_axis**2
create_axis = empty_canvas.add_axes([0, 0, 1, 1]) 

create_axis.plot(x_axis, y_axis, 'b')



create_axis = empty_canvas.add_axes([0, 0, 1, 1]) #x_index,y_index,width,height

create_axis.plot(x_axis, y_axis, 'b', label = "x = x**2")
create_axis.plot(x_axis, y_axis+(x_axis*3), 'g', label = "x = y+(x*3)")

create_axis.legend(loc=0)
```

```python
import matplotlib.pyplot as plt

data = {'Batch 1': 10, 'Batch 2': 15, 'Batch3': 5, 'Batch 4': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1,3, figsize=(10, 3))#1 row and 3 columns indexing

axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)

fig.suptitle('Subplots')
```
